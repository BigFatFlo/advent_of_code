```python
f = open("input.txt", "r")

total = 0
text_digits = {"one":'1', "two":'2', "three":'3', "four":'4', "five":'5', "six":'6', "seven":'7', "eight":'8', "nine":'9'}

def get_digit(line: str, i: int)->str:
    n = len(line)
    if line[i].isdigit():
        return line[i]
    elif i+3 <= n and line[i:i+3] in text_digits:
            return text_digits[line[i:i+3]]
    elif i+4 <= n and line[i:i+4] in text_digits:
            return text_digits[line[i:i+4]]
    elif i+5 <= n and line[i:i+5] in text_digits:
            return text_digits[line[i:i+5]]
    else:
        return ''


for line in f:
    first_digit = -1
    last_digit = -1
    n = len(line)
    for i in range(n):
        c = get_digit(line, i)
        if c != '':
            first_digit = c
            break

    for i in range(n):
        c = get_digit(line, n-1-i)
        if c != '':
            last_digit = c
            break

    result = int(first_digit + last_digit)
    print(result)
    total += result

print(total)

```
