```python
from queue import Queue
from tabulate import tabulate

f = open("input.txt", "r")

grid = []

for line in f:
    grid.append([c for c in line.replace('\n', '')])

height = len(grid)
width = len(grid[0])

starting_position = [0]
final_position = [height-1]

for j in range(width):
    if grid[0][j] == '.':
        starting_position.append(j)
    if grid[height-1][j] == '.':
        final_position.append(j)
    if len(starting_position) == 2 and len(final_position) == 2:
        break

print(tabulate(grid))
print(starting_position, final_position)

directions = {'<': (0, -1), '>': (0, 1), 'v': (1, 0), '^': (-1, 0)}

def takeStep(heads,paths):
    (s, path_index, (i, j)) = heads.get()
    current_tile = grid[i][j]
    paths[path_index].add((i, j))
    if [i, j] == final_position:
        max_steps = s
        return True, max_steps
    new_positions = []
    match current_tile:
        case '.':
            possible_directions = directions.values()
        case '<' | '>' | 'v' | '^':
            possible_directions = [directions[current_tile]]
        case _:
            possible_directions = []
    for (ii, jj) in possible_directions:
        (new_i, new_j) = (i+ii, j+jj)
        if 0 <= new_i < height and 0 <= new_j < width:
            if grid[new_i][new_j] != '#' and (new_i, new_j) not in paths[path_index]:
                new_positions.append((new_i, new_j))
    for pos_index, p in enumerate(new_positions):
        if pos_index == 0:
            heads.put((s+1, path_index, p))
        else:
            paths.append(set())
            paths[-1].update(paths[path_index])
            heads.put((s+1, len(paths)-1, p))
    return False, 0

heads = Queue()
heads.put((0, 0, tuple(starting_position)))
paths = [set()]

reached_end = False
max_steps_list = []
while not heads.empty():
    reached_end, max_steps = takeStep(heads, paths)
    if reached_end:
        max_steps_list.append(max_steps)

print(max_steps_list)
print(max(max_steps_list))
```

```python
from collections import defaultdict
from os import path
from queue import Queue

from numpy import cross

f = open("input.txt", "r")

grid = []

for line in f:
    grid.append([c for c in line.replace('\n', '')])

height = len(grid)
width = len(grid[0])

starting_position = [0]
final_position = [height-1]

for j in range(width):
    if grid[0][j] == '.':
        starting_position.append(j)
    if grid[height-1][j] == '.':
        final_position.append(j)
    if len(starting_position) == 2 and len(final_position) == 2:
        break


for i in range(height):
    for j in range(width):
        if grid[i][j] != '#':
            grid[i][j] = '.'

starting_position = tuple(starting_position)
final_position = tuple(final_position)

directions = {(0, -1), (0, 1), (1, 0), (-1, 0)}

nodes = set([starting_position, final_position])

print(tabulate(grid))

for i in range(height):
    for j in range(width):
        if grid[i][j] != '.':
            continue
        nb_of_reachable_tiles = 0
        for (ii, jj) in directions:
            (new_i, new_j) = (i+ii, j+jj)
            if 0 <= new_i < height and 0 <= new_j < width:
                if grid[new_i][new_j] != '#':
                    nb_of_reachable_tiles+=1
                if nb_of_reachable_tiles > 2:
                    nodes.add((i, j))

print(nodes)
for c in nodes:
    grid[c[0]][c[1]] = 'O'

crossroads = {}

for node in nodes:
    crossroads[node] = {}
    visited = set()

    queue = Queue()
    queue.put((node, 0))
    while not queue.empty():
        ((i, j), steps) = queue.get()
        visited.add((i, j))
        
        steps += 1
        for (ii, jj) in directions:
            (new_i, new_j) = (i+ii, j+jj)
            if 0 <= new_i < height and 0 <= new_j < width and grid[new_i][new_j] != '#' and (new_i, new_j) not in visited:
                if (new_i, new_j) in nodes:
                    crossroads[node][(new_i, new_j)] = steps
                else:
                    queue.put(((new_i, new_j), steps))

for node in crossroads:
    print(node, crossroads[node])

def maxStepsFromTo(s, e, crossroads, visited):
    path_lengths = []
    if s == e:
        return [0]
    visited.add(s)
    for possible_start, steps in crossroads[s].items():
        if possible_start in visited:
            continue
        for path_len in maxStepsFromTo(possible_start, e, crossroads, visited):
            path_lengths.append(steps + path_len)
    visited.remove(s)
    return path_lengths

paths = maxStepsFromTo(starting_position, final_position, crossroads, set())
print(max(paths))
```
