```python
from tabulate import tabulate

f = open("input.txt", "r")

grid = []

for line in f:
    grid.append([c for c in line.replace('\n', '')])

height = len(grid)
width = len(grid[0])

print(tabulate(grid))

def newDir(dir, i, j, s):
    match dir:
        case 'r':
            if j + 1 < width:
                s.add((i, j+1, 'r'))
        case 'l':
            if j - 1 >= 0:
                s.add((i, j-1, 'l'))
        case 'd':
            if i + 1 < height:
                s.add((i+1, j, 'd'))
        case 'u':
            if i - 1 >= 0:
                s.add((i-1, j, 'u'))

def energize(start):
    energized = [[0 for j in range(width)] for i in range(height)]
    heads = {start}
    new_energized = 0
    total_energized = 0
    loops_at_0 = 0
    handled_heads = set()
    while loops_at_0 < 20:
        if new_energized == 0:
            loops_at_0+=1
        else:
            loops_at_0 = 0
        new_heads = set()
        new_energized = 0
        for (i, j, dir) in heads:
            if (i, j, dir) in handled_heads:
                continue
            else:
                handled_heads.add((i, j, dir))
            if energized[i][j] == 0:
                new_energized +=1
                total_energized += 1
            energized[i][j] = 1
            c = grid[i][j]
            match dir, c:
                case 'r' | 'l', '|':
                    newDir('u', i, j, new_heads)
                    newDir('d', i, j, new_heads)
                case 'r', '/':
                    newDir('u', i, j, new_heads)
                case 'r', '\\':
                    newDir('d', i, j, new_heads)
                case 'r', '.' | '-':
                    newDir('r', i, j, new_heads)
                case 'l', '/':
                    newDir('d', i, j, new_heads)
                case 'l', '\\':
                    newDir('u', i, j, new_heads)
                case 'l', '.' | '-':
                    newDir('l', i, j, new_heads)
                case 'd' | 'u', '-':
                    newDir('l', i, j, new_heads)
                    newDir('r', i, j, new_heads)
                case 'd', '/':
                    newDir('l', i, j, new_heads)
                case 'd', '\\':
                    newDir('r', i, j, new_heads)
                case 'd', '.' | '|':
                    newDir('d', i, j, new_heads)
                case 'u', '/':
                    newDir('r', i, j, new_heads)
                case 'u', '\\':
                    newDir('l', i, j, new_heads)
                case 'u', '.' | '|':
                    newDir('u', i, j, new_heads)
                case _:
                    print("unexpected")
        heads = new_heads
    print(total_energized)
    return total_energized

max_energized = 0

for j in range(width):
    max_energized = max(energize((0, j, 'd')), max_energized)
for j in range(width):
    max_energized = max(energize((height-1, j, 'u')), max_energized)
for i in range(height):
    max_energized = max(energize((i, 0, 'r')), max_energized)
for i in range(height):
    max_energized = max(energize((i, width-1, 'l')), max_energized)

print(max_energized)
```
