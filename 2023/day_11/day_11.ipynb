```python
from tabulate import tabulate
from collections import deque

f = open("input.txt", "r")

map = []

for line in f:
    map.append(list(line.replace('\n', '')))

print(tabulate(map))

height = len(map)
width = len(map[0])

empty_rows = set([row for row in range(height)])
empty_cols = set([col for col in range(width)])

for i in range(height):
    for j in range(width):
        if map[i][j] == '#':
            try:
                empty_rows.remove(i)
            except KeyError:
                pass
            try:
                empty_cols.remove(j)
            except KeyError:
                pass

ex_map = [['.' for j in range(width + len(empty_cols))] for i in range(height + len(empty_rows))]

i_m = 0
i_em = 0
j_m = 0
j_em = 0
while i_m < height:
    j_m = 0
    j_em = 0
    while j_m < width:
        ex_map[i_em][j_em] = map[i_m][j_m]
        if j_m in empty_cols:
            j_em += 1
        j_m += 1
        j_em += 1
    if i_m in empty_rows:
        i_em += 1
    i_m += 1
    i_em += 1

height = i_em
width = j_em

map = ex_map

print(tabulate(map))

galaxies = []
for i in range(height):
    for j in range(width):
        if map[i][j] == '#':
            galaxies.append((i, j))

galaxies = deque(galaxies)
shortest_paths = []
while len(galaxies) != 0:
    src = galaxies.popleft()
    dist = []
    for dest in galaxies:
        dist.append(abs(dest[1] - src[1]) + abs(dest[0] - src[0]))
    shortest_paths.append(dist)

print(sum(sum(dist) for dist in shortest_paths))

```

```python
from tabulate import tabulate
from collections import deque

f = open("input.txt", "r")

map = []

for line in f:
    map.append(list(line.replace('\n', '')))

print(tabulate(map))

height = len(map)
width = len(map[0])

# For each galaxy, we'll keep track of its original coordinates and its coordinates after expansion
galaxy_grid = [[[] for j in range(width)] for i in range(height)]
empty_rows = set([row for row in range(height)])
empty_cols = set([col for col in range(width)])

for i in range(height):
    for j in range(width):
        if map[i][j] == '#':
            galaxy_grid[i][j].append((i, j))
            try:
                empty_rows.remove(i)
            except KeyError:
                pass
            try:
                empty_cols.remove(j)
            except KeyError:
                pass

print(galaxy_grid)

i_m = 0
i_em = 0
j_m = 0
j_em = 0
while i_m < height:
    j_m = 0
    j_em = 0
    while j_m < width:
        if map[i_m][j_m] == '#':
            # Here we update the galaxy at i_m and j_m in the 
            # original grid with its coordinates in the expanded grid
            galaxy_grid[i_m][j_m].append((i_em, j_em))
        if j_m in empty_cols:
            j_em += 999999
        j_m += 1
        j_em += 1
    if i_m in empty_rows:
        i_em += 999999
    i_m += 1
    i_em += 1

height = i_em
width = j_em

map = ex_map

print(galaxy_grid)

# We're only interested in the galaxies in the expanded grid
galaxies = []
for i in range(len(galaxy_grid)):
    for j in range(len(galaxy_grid[0])):
        gal = galaxy_grid[i][j]
        if len(gal) == 2:
            galaxies.append(gal[1])

print(galaxies)
galaxies = deque(galaxies)
shortest_paths = []
while len(galaxies) != 0:
    src = galaxies.popleft()
    dist = []
    for dest in galaxies:
        dist.append(abs(dest[1] - src[1]) + abs(dest[0] - src[0]))
    shortest_paths.append(dist)

print(sum(sum(dist) for dist in shortest_paths))

```
