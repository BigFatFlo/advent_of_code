```python
from tabulate import tabulate

f = open("input.txt", "r")

grid = []

for line in f:
    grid.append([c for c in line.replace('\n', '')])

height = len(grid)
width = len(grid[0])

def tiltNorth(grid):
    for j in range(width):
        nb_of_rocks = 0
        i = 0
        start = i
        while i < height:
            if grid[i][j] == 'O':
                grid[i][j] = '.'
                nb_of_rocks += 1
            if grid[i][j] == '#' or i == height-1:
                for ii in range(start, start + nb_of_rocks):
                    grid[ii][j] = 'O'
                start = i+1
                nb_of_rocks = 0
            i+=1

def tiltSouth(grid):
    for j in range(width):
        nb_of_rocks = 0
        i = height - 1
        start = i
        while i >= 0:
            if grid[i][j] == 'O':
                grid[i][j] = '.'
                nb_of_rocks += 1
            if grid[i][j] == '#' or i == 0:
                for ii in range(start, start - nb_of_rocks, -1):
                    grid[ii][j] = 'O'
                start = i-1
                nb_of_rocks = 0
            i-=1

def tiltWest(grid):
    for i in range(height):
        start = 0
        nb_of_rocks = 0
        j = 0
        while j < width:
            if grid[i][j] == 'O':
                grid[i][j] = '.'
                nb_of_rocks += 1
            if grid[i][j] == '#' or j == width-1:
                for jj in range(start, start + nb_of_rocks):
                    grid[i][jj] = 'O'
                start = j+1
                nb_of_rocks = 0
            j+=1

def tiltEast(grid):
    for i in range(height):
        nb_of_rocks = 0
        j = width - 1
        start = j
        while j >= 0:
            if grid[i][j] == 'O':
                grid[i][j] = '.'
                nb_of_rocks += 1
            if grid[i][j] == '#' or j == 0:
                for jj in range(start, start - nb_of_rocks, -1):
                    grid[i][jj] = 'O'
                start = j-1
                nb_of_rocks = 0
            j-=1

def cycle(grid):
    tiltNorth(grid)
    tiltWest(grid)
    tiltSouth(grid)
    tiltEast(grid)

def load(grid):
    total = 0
    for i in range(height):
        for j in range(width):
            if grid[i][j] == 'O':
                total+= height-i
    return total

totals = [] 
totals.append(load(grid))
for _ in range(500):
    cycle(grid)
    totals.append(load(grid))

print(tabulate(grid))

# Floyd's tortoise and hare algo to find cycle

x = 0
t = x+1
h = x+2

def foundCycle(totals, t, h):
    for i in range(10):
        if totals[t+i] != totals[h+i]:
            return False
    return True

while not foundCycle(totals, t, h):
    t+=1
    h+=2

print(t)

cycle_start = x
t = x

while totals[t] != totals[h]:
    t+=1
    h+=1
    cycle_start+=1

print(cycle_start)
print(totals)
print(totals[cycle_start:])

cycle_length=1
t+=1
while not foundCycle(totals, t, h):
    t+=1
    cycle_length+=1

print(cycle_length)

cycle_start_value = totals[cycle_start]
i = 0
while totals[i] != cycle_start_value:
    if not foundCycle(totals, i, i+cycle_length):
        i+=1

cycle_start = i

print(cycle_start)
print(totals[cycle_start:])

print(totals[cycle_start+((1000000000 - cycle_start)%cycle_length)])
```
