```python
from tabulate import tabulate

f = open("input.txt", "r")

patterns = []
new_pattern = []

for line in f:
    if line == '\n':
        patterns.append(new_pattern)
        new_pattern = []
    else:
        new_pattern.append(line.replace('\n', ''))

total = 0

for pattern in patterns:
    mirror_found = False
    height = len(pattern)
    width = len(pattern[0])

    # To each possible vertical mirror we associate the number of mismatches
    # Position j means the mirror is between column j-1 and j (starts at 0)
    potential_vertical_mirror = {}
    for mirror in range(1, width):
        potential_vertical_mirror[mirror] = 0
    i = 0
    while i < height:
        for mirror in range(1, width):
            for j in range(1, min(mirror+1, width-mirror+1)):
                if pattern[i][mirror-j] != pattern[i][mirror+j-1]:
                    potential_vertical_mirror[mirror]+=1
        i+=1

    for mirror, mismatches in potential_vertical_mirror.items():
        if mismatches == 1:
            total += mirror
            mirror_found = True
            break
    if mirror_found:
        continue
    
    # To each possible horizontal mirror we associate the number of mismatches
    # Position i means the mirror is between row i-1 and i (starts at 0)
    potential_horizontal_mirror = {}
    for mirror in range(1, height):
        potential_horizontal_mirror[mirror] = 0
    j = 0
    while j < width:
        for mirror in range(1, height):
            for i in range(1, min(mirror+1, height-mirror+1)):
                if pattern[mirror-i][j] != pattern[mirror+i-1][j]:
                    potential_horizontal_mirror[mirror]+=1
        j+=1

    for mirror, mismatches in potential_horizontal_mirror.items():
        if mismatches == 1:
            total += 100*mirror
            mirror_found = True
            break
    if mirror_found:
        continue

print(total)


```
