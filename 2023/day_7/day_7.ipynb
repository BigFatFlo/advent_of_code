```python
f = open("input.txt", "r")
kind_rank = {"high": 0, "pair": 1, "2pairs": 2, "3oak": 3, "full": 4, "4oak": 5, "5oak": 6}
card_rank = ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]

class Hand:
    def __init__(self, hand, bid):
        self.hand = hand
        self.bid = bid
        self.setKind()

    def __gt__(self, other):
        if self.kind != other.kind:
            return kind_rank[self.kind] > kind_rank[other.kind]
        else:
            i = 0
            while self.hand[i] == other.hand[i] and i < len(self.hand):
                i += 1
            return card_rank.index(self.hand[i]) < card_rank.index(other.hand[i])
        
    def setKind(self):
        hand = self.hand
        nb_of_pairs = 0
        nb_of_3oak = 0
        nb_of_4oak = 0
        nb_of_5oak = 0
        while len(hand) > 0:
            c = hand[0]
            match (hand.count(c)):
                case 2:
                    nb_of_pairs += 1
                case 3:
                    nb_of_3oak += 1
                case 4:
                    nb_of_4oak += 1
                case 5:
                    nb_of_5oak += 1
                case _:
                    pass
            hand = hand.replace(c, "")
        if nb_of_5oak == 1:
            self.kind = "5oak"
        elif nb_of_4oak == 1:
            self.kind = "4oak"
        elif nb_of_3oak == 1:
            if nb_of_pairs == 0:
                self.kind = "3oak"
            else:
                self.kind = "full"
        elif nb_of_pairs == 2:
            self.kind = "2pairs"
        elif nb_of_pairs == 1:
            self.kind = "pair"
        else:
            self.kind = "high"


hands = []

for line in f:
    fields = line.replace("\n", "").split(" ")
    hand = fields[0]
    bid = int(fields[1])
                
    hands.append(Hand(hand, bid))

sorted_hands = sorted(hands)
total = 0
for index, hand in enumerate(sorted_hands):
    total += (index+1) * hand.bid

print(total)
```

```python
f = open("input.txt", "r")

card_rank = ["A", "K", "Q", "T", "9", "8", "7", "6", "5", "4", "3", "2", "J"]

class JokerHand:
    def __init__(self, hand, bid):
        self.hand = hand
        self.bid = bid
        self.setKind()

    def __gt__(self, other):
        if self.kind != other.kind:
            return kind_rank[self.kind] > kind_rank[other.kind]
        else:
            i = 0
            while self.hand[i] == other.hand[i] and i < len(self.hand):
                i += 1
            return card_rank.index(self.hand[i]) < card_rank.index(other.hand[i])
        
    def setKind(self):
        hand = self.hand
        nb_of_jokers = hand.count('J')
        if nb_of_jokers == 5:
            self.kind = "5oak"
            return
        possible_kinds = []
        for i in range(len(hand)):
            c = hand[i]
            if (c == 'J'):
                continue
            nb_of_pairs = 0
            nb_of_3oak = 0
            nb_of_4oak = 0
            nb_of_5oak = 0
            temp_hand = hand.replace('J', c)
            while len(temp_hand) > 0:
                c = temp_hand[0]
                match (temp_hand.count(c)):
                    case 2:
                        nb_of_pairs += 1
                    case 3:
                        nb_of_3oak += 1
                    case 4:
                        nb_of_4oak += 1
                    case 5:
                        nb_of_5oak += 1
                    case _:
                        pass
                temp_hand = temp_hand.replace(c, "")
            if nb_of_5oak == 1:
                possible_kinds.append("5oak")
            elif nb_of_4oak == 1:
                possible_kinds.append("4oak")
            elif nb_of_3oak == 1:
                if nb_of_pairs == 0:
                    possible_kinds.append("3oak")
                else:
                    possible_kinds.append("full")
            elif nb_of_pairs == 2:
                possible_kinds.append("2pairs")
            elif nb_of_pairs == 1:
                possible_kinds.append("pair")
            else:
                possible_kinds.append("high")
        self.kind = possible_kinds[0]
        for kind in possible_kinds:
            if kind_rank[kind] > kind_rank[self.kind]:
                self.kind = kind

hands = []

for line in f:
    fields = line.replace("\n", "").split(" ")
    hand = fields[0]
    bid = int(fields[1])
                
    hands.append(JokerHand(hand, bid))

sorted_hands = sorted(hands)
total = 0
for index, hand in enumerate(sorted_hands):
    total += (index+1) * hand.bid

print(total)
```
