```python
input = "hxbxwxba"
examples = ["abcdefgh", "ghijklmn"]

straights = [chr(i)+chr(i+1)+chr(i+2) for i in range(ord('a'), ord('z')-1)]
forbidden_letters = ['i', 'o', 'l']

def onestep(s, k):
    c = s[k]
    wrap = False
    if c == 'z':
        new_c = 'a'
        wrap = True
    else:
        new_c = chr(ord(c)+1)
        if new_c in forbidden_letters:
            new_c = chr(ord(new_c)+1)
    return wrap, s[:k]+new_c+s[k+1:]

def includesStraight(s):
    res = False
    for straight in straights:
        res = straight in s
        if res:
            break
    return res

def includeConfusingLetters(s):
    return 'i' in s or 'o' in s or 'l' in s

def includeTwoPairs(s):
    nb_of_pairs = 0
    found_pairs = set()
    i = 0
    while i < len(s)-1:
        c = s[i]
        next_c = s[i+1]
        if next_c == c:
            pair = c+next_c
            if pair not in found_pairs:
                found_pairs.add(pair)
                nb_of_pairs +=1
            i+=2
        else:
            i+=1
    return nb_of_pairs >= 2

def isValid(s):
    return includesStraight(s) and not includeConfusingLetters(s) and includeTwoPairs(s)

def increment(s):
    l = len(s)
    for i in range(l):
        wrap, s = onestep(s, l-1-i)
        if not wrap:
            break
    return s

def skipForbiddenCharacter(s):
    # If there is a forbidden character in the string, we can
    # skip all the steps until it's gone
    l = len(s)
    for i in range(l):
        c = s[i]
        if c in forbidden_letters:
            s = s[:i]+chr(ord(c)+1)
            for k in range(l-len(s)):
                s+='a'
            break
    return s

def findNextValidPassword(s):
    s = skipForbiddenCharacter(s)
    while not isValid(s):
        s = increment(s)
    return s

for example in examples:
    print(findNextValidPassword(example))

next = findNextValidPassword(input)
print(next)
next = findNextValidPassword(increment(next))
print(next)

```
