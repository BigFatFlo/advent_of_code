```python
from collections import defaultdict

f = open("input.txt", 'r')

instr_list = []

for line in f:
    instr_list.append(list(line.replace('\n','')))

print(instr_list)

dirs = {'<': (0, -1), '>': (0, 1), '^': (-1, 0), 'v': (1, 0)}

for instr in instr_list:
    visited = defaultdict(int)
    i, j = 0, 0
    visited[(i, j)] += 1
    for c in instr:
        dir = dirs[c]
        i, j = i+dir[0], j+dir[1]
        visited[(i, j)] += 1
    print(visited)
    print(len(visited))


for instr in instr_list:
    visited = defaultdict(int)
    santa_i, santa_j = 0, 0
    robot_i, robot_j = 0, 0
    visited[(santa_i, santa_j)] += 1
    visited[(robot_i, robot_j)] += 1
    santa = True
    for c in instr:
        dir = dirs[c]
        if santa:
            santa_i, santa_j = santa_i+dir[0], santa_j+dir[1]
            visited[(santa_i, santa_j)] += 1
        else:
            robot_i, robot_j = robot_i+dir[0], robot_j+dir[1]
            visited[(robot_i, robot_j)] += 1
        santa = not santa
    print(visited)
    print(len(visited))


```
