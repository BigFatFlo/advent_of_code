```python
f = open("input.txt", "r")
input_txt = f.read()
print(input_txt)

input_list = list(input_txt)
print(input_list)

floor = 0

for i in range(len(input_list)):
    c = input_list[i]
    if c == '(':
        floor += 1
    elif c == ')':
        floor -= 1
    if floor < 0:
        break

print(floor)
print(i+1)

```
