```python
from itertools import product

example_player_stats = {"hp": 8, "dmg":5, "armor":5}
example_boss_stats = {"hp": 12, "dmg":7, "armor":2}

def playGame(player_stats, boss_stats):
    while True:
        dmg_dealt = max(1, player_stats["dmg"]-boss_stats["armor"])
        boss_stats["hp"]-=dmg_dealt
        if boss_stats["hp"] <= 0:
            return "Player Wins"
        dmg_dealt = max(1, boss_stats["dmg"]-player_stats["armor"])
        player_stats["hp"]-=dmg_dealt
        if player_stats["hp"] <= 0:
            return "Boss Wins"

def fillShop():
    shop_file = open("shop.txt", 'r')
    shop = {"Weapons": {}, "Armor": {}, "Rings": {}}
    category = ""
    for line in shop_file:
        if line.startswith("Weapons"):
            category = "Weapons"
            continue
        if line.startswith("Armor"):
            category = "Armor"
            continue
        if line.startswith("Rings"):
            category = "Rings"
            continue
        if line == '\n':
            continue
        while line.find("  ") != -1:
            line = line.replace("  ", " ").replace('\n', '')
        desc = line.split(" ")
        if category in ["Weapons", "Armor"]:
            name = desc[0]
            cost = int(desc[1])
            dmg = int(desc[2])
            armor = int(desc[3])
        else:
            name = desc[0]+desc[1]
            cost = int(desc[2])
            dmg = int(desc[3])
            armor = int(desc[4])
        shop[category][name] = {"cost": cost, "dmg": dmg, "armor": armor}
    return shop

def fillBossStats():
    stats_file = open("stats.txt", 'r')
    boss_stats = {"hp": 0, "dmg":0, "armor":0}
    for line in stats_file:
        if line.startswith("Hit Points"):
            boss_stats["hp"]=int(line.replace(' ', '').replace('\n', '').split(':')[1])
        if line.startswith("Damage"):
            boss_stats["dmg"]=int(line.replace(' ', '').replace('\n', '').split(':')[1])
        if line.startswith("Armor"):
            boss_stats["armor"]=int(line.replace(' ', '').replace('\n', '').split(':')[1])
    return boss_stats

def getPlayerStats(weapon, armor, ring1, ring2):
    stats = {"hp": 100, "dmg":0, "armor":0, "cost": 0}
    stats["dmg"] += weapon["dmg"]
    stats["cost"] += weapon["cost"]
    if armor != None:
        stats["armor"] += armor["armor"]
        stats["cost"] += armor["cost"]
    if ring1 != None:
        stats["dmg"] += ring1["dmg"]
        stats["armor"] += ring1["armor"]
        stats["cost"] += ring1["cost"]
    if ring2 != None:
        stats["dmg"] += ring2["dmg"]
        stats["armor"] += ring2["armor"]
        stats["cost"] += ring2["cost"]
    return stats

def getInventories(shop):
    inventories = []
    for weapon, armor, ring1, ring2 in product(shop["Weapons"].values(), [None, *shop["Armor"].values()], [None, *shop["Rings"].values()], [None, *shop["Rings"].values()]):
        inventories.append((weapon, armor, ring1, ring2))
    return inventories

def part1():
    shop = fillShop()
    inventories = getInventories(shop)
    costs = []
    for weapon, armor, ring1, ring2 in inventories:
        player_stats = getPlayerStats(weapon, armor, ring1, ring2)
        boss_stats = fillBossStats()
        if playGame(player_stats, boss_stats) == "Player Wins":
            costs.append(player_stats["cost"])
    print(min(costs))

def part2():
    shop = fillShop()
    inventories = getInventories(shop)
    costs = []
    for weapon, armor, ring1, ring2 in inventories:
        player_stats = getPlayerStats(weapon, armor, ring1, ring2)
        boss_stats = fillBossStats()
        if playGame(player_stats, boss_stats) == "Boss Wins":
            costs.append(player_stats["cost"])
    print(max(costs))

print(playGame(example_player_stats, example_boss_stats))
part1()
part2()
```
